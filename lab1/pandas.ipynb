# Fictitious Names

##  Create the 3 DataFrames based on the following raw data

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
first = pd.DataFrame(raw_data_1)
second = pd.DataFrame(raw_data_2)
third = pd.DataFrame(raw_data_3)
```

## Assign each to a variable called data1, data2, data3

```python
data1 = first
data2 = second
data3 = third
```

## Join the two dataframes along rows and assign all_data

```python
all_data = pd.concat([data1, data2], axis=0)

all_data
```

## Join the two dataframes along columns and assing to all_data_col

```python
all_data_col = pd.concat([data1, data2], axis=1)

all_data_col
```

## Merge all_data and data3 along the subject_id value

```python
result = pd.merge(all_data, data3, on="subject_id")

result
```

## Merge only the data that has the same 'subject_id' on both data1 and data2

```python
result = pd.merge(data1, data2, on="subject_id", how="inner")

result
```

## Merge all values in data1 and data2, with matching records from both sides where available.

```python
result = pd.merge(data1, data2, on="subject_id", how="outer")

result
```

# Euro 12

## Import data from datasets/euro12.csv

```python
euro12 = pd.read_csv("../data/euro12.csv")

euro12
```

##  Calculate the mean Yellow Cards given per Team

```python
result = euro12.groupby("Team").aggregate({"Yellow Cards": ["mean"]})

result
```

## Which teams scored more than 6 goals?

```python
result = euro12[euro12.Goals > 6][["Team", "Goals"]].reset_index(drop=True)

result
```

## Select the teams that start with G

```python
result = euro12[euro12.Team.str.startswith("G")][["Team"]].reset_index(drop=True)

result
```

## Present only the Shooting Accuracy from England, Italy and Russia

```python
result = euro12[euro12.Team.isin(["England", "Italy", "Russia"])][["Team", "Shooting Accuracy"]].reset_index(drop=True)

result
```

# Drinks

## Read and save data from datasets/drinks.txt

```python
drinks = pd.read_csv("../data/drinks.csv")

drinks
```

## Which continent drinks more beer on average?

```python
result = drinks.groupby("continent").mean()["beer_servings"].idxmax()
print("Contintent which drinks more beer on average: " + str(result))
```

## For each continent print the statistics for wine consumption.

```python
result = drinks.groupby("continent").aggregate("wine_servings").describe()

result
```

##  Print the mean, min and max values for spirit consumption per continent.

```python
result = drinks.groupby("continent").aggregate({"spirit_servings": ["mean", "min", "max"]})

result
```
